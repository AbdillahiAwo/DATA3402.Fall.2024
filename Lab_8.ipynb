```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, classification_report

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lda = LinearDiscriminantAnalysis()

lda.fit(X_train, y_train)

y_pred = lda.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of LDA on Iris dataset:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

print("\nExplained variance ratio:", lda.explained_variance_ratio_)
```

#### Part a

Compare ROC curves computed on the test versus training samples, in a single plot. Do you see a bias?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

iris = load_iris()
X = iris.data
y = (iris.target == 2).astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

logreg = LogisticRegression(max_iter=200, solver='liblinear')
logreg.fit(X_train, y_train)

y_train_proba = logreg.predict_proba(X_train)[:, 1]
y_test_proba = logreg.predict_proba(X_test)[:, 1]

fpr_train, tpr_train, _ = roc_curve(y_train, y_train_proba)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)

auc_train = auc(fpr_train, tpr_train)
auc_test = auc(fpr_test, tpr_test)

plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, color='blue', label=f'Train ROC (AUC = {auc_train:.2f})')
plt.plot(fpr_test, tpr_test, color='red', linestyle='--', label=f'Test ROC (AUC = {auc_test:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Training vs Test Data')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import roc_curve, auc

iris = load_iris()
X, y = iris.data, (iris.target == 2).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train_raw = X_train
X_test_raw = X_test

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

X_train_combined = np.hstack((X_train_raw, X_train_pca))
X_test_combined = np.hstack((X_test_raw, X_test_pca))

lda_raw = LinearDiscriminantAnalysis()
lda_pca = LinearDiscriminantAnalysis()
lda_combined = LinearDiscriminantAnalysis()

lda_raw.fit(X_train_raw, y_train)
lda_pca.fit(X_train_pca, y_train)
lda_combined.fit(X_train_combined, y_train)

y_proba_raw = lda_raw.predict_proba(X_test_raw)[:, 1]
y_proba_pca = lda_pca.predict_proba(X_test_pca)[:, 1]
y_proba_combined = lda_combined.predict_proba(X_test_combined)[:, 1]

fpr_raw, tpr_raw, _ = roc_curve(y_test, y_proba_raw)
fpr_pca, tpr_pca, _ = roc_curve(y_test, y_proba_pca)
fpr_combined, tpr_combined, _ = roc_curve(y_test, y_proba_combined)

auc_raw = auc(fpr_raw, tpr_raw)
auc_pca = auc(fpr_pca, tpr_pca)
auc_combined = auc(fpr_combined, tpr_combined)

plt.figure(figsize=(10, 6))

plt.plot(fpr_raw, tpr_raw, label=f'Raw (AUC = {auc_raw:.2f})', color='blue')
plt.plot(fpr_pca, tpr_pca, label=f'PCA Features (AUC = {auc_pca:.2f})', color='red', linestyle='--')
plt.plot(fpr_combined, tpr_combined, label=f'Raw + PCA Features (AUC = {auc_combined:.2f})', color='green')

plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Raw, Features, and Raw + Features')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc

iris = load_iris()
X, y = iris.data, (iris.target == 2).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

logreg = LogisticRegression(max_iter=200, solver='liblinear', random_state=42)
dtree = DecisionTreeClassifier(max_depth=5, random_state=42)
rforest = RandomForestClassifier(n_estimators=100, random_state=42)

logreg.fit(X_train, y_train)
dtree.fit(X_train, y_train)
rforest.fit(X_train, y_train)

y_proba_logreg = logreg.predict_proba(X_test)[:, 1]
y_proba_dtree = dtree.predict_proba(X_test)[:, 1]
y_proba_rforest = rforest.predict_proba(X_test)[:, 1]

fpr_logreg, tpr_logreg, _ = roc_curve(y_test, y_proba_logreg)
fpr_dtree, tpr_dtree, _ = roc_curve(y_test, y_proba_dtree)
fpr_rforest, tpr_rforest, _ = roc_curve(y_test, y_proba_rforest)

auc_logreg = auc(fpr_logreg, tpr_logreg)
auc_dtree = auc(fpr_dtree, tpr_dtree)
auc_rforest = auc(fpr_rforest, tpr_rforest)

plt.figure(figsize=(10, 6))

plt.plot(fpr_logreg, tpr_logreg, label=f'Logistic Regression (AUC = {auc_logreg:.2f})', color='blue')
plt.plot(fpr_dtree, tpr_dtree, label=f'Decision Tree (AUC = {auc_dtree:.2f})', color='red', linestyle='--')
plt.plot(fpr_rforest, tpr_rforest, label=f'Random Forest (AUC = {auc_rforest:.2f})', color='green')

plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Logistic Regression, Decision Tree, and Random Forest')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def compare_classifier_performance(classifier, X_train, X_test, y_train, y_test, model_name="Classifier"):
    """
    Train a classifier, predict probabilities, and plot ROC curves for both training and testing sets.

    Parameters:
    -----------
    classifier: Scikit-learn classifier instance (must have fit() and predict_proba() methods)
    X_train: Training feature data
    X_test: Testing feature data
    y_train: Training labels
    y_test: Testing labels
    model_name: Name of the classifier (used in plots and output)

    Returns:
    --------
    None (displays the plot and prints AUC scores)
    """

    classifier.fit(X_train, y_train)

    y_proba_train = classifier.predict_proba(X_train)[:, 1]
    y_proba_test = classifier.predict_proba(X_test)[:, 1]


    fpr_train, tpr_train, _ = roc_curve(y_train, y_proba_train)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_proba_test)

    auc_train = auc(fpr_train, tpr_train)
    auc_test = auc(fpr_test, tpr_test)

    plt.figure(figsize=(10, 6))

    plt.plot(fpr_train, tpr_train, label=f'{model_name} (Train AUC = {auc_train:.2f})', color='blue')
    plt.plot(fpr_test, tpr_test, label=f'{model_name} (Test AUC = {auc_test:.2f})', color='red', linestyle='--')

    plt.plot([0, 1], [0, 1], color='grey', linestyle='--')

    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title(f'ROC Curve for {model_name}')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

    print(f"AUC for {model_name} (Training set): {auc_train:.2f}")
    print(f"AUC for {model_name} (Testing set): {auc_test:.2f}")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

iris = load_iris()
X, y = iris.data, (iris.target == 2).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

logreg = LogisticRegression(max_iter=200, solver='liblinear', random_state=42)
dtree = DecisionTreeClassifier(max_depth=5, random_state=42)
rforest = RandomForestClassifier(n_estimators=100, random_state=42)

print("\n--- Logistic Regression ---")
compare_classifier_performance(logreg, X_train, X_test, y_train, y_test, model_name="Logistic Regression")

print("\n--- Decision Tree ---")
compare_classifier_performance(dtree, X_train, X_test, y_train, y_test, model_name="Decision Tree")

print("\n--- Random Forest ---")
compare_classifier_performance(rforest, X_train, X_test, y_train, y_test, model_name="Random Forest")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def compute_maximal_significance(classifier, X_train, X_test, y_train, y_test, model_name="Classifier"):
    """
    Train a classifier, classify events as signal/background, and compute the maximal significance for Lab 5 scenarios.

    Parameters:
    -----------
    classifier: Scikit-learn classifier instance (must have fit() and predict() methods)
    X_train: Training feature data
    X_test: Testing feature data
    y_train: Training labels (1 for signal, 0 for background)
    y_test: Testing labels (1 for signal, 0 for background)
    model_name: Name of the classifier (used in plots and output)

    Returns:
    --------
    sigma_s (float): The computed maximal significance
    N_S (int): Number of signal events
    N_B (int): Number of background events
    """

    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)

    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    N_S = tp
    N_B = fp

    if (N_S + N_B) == 0:
        sigma_S = 0
    else:

        sigma_S = N_S / np.sqrt(N_S + N_B)

    print(f"\n--- {model_name} ---")
    print(f"Number of signal events (N_S): {N_S}")
    print(f"Number of background events (N_B): {N_B}")
    print(f"Maximal Significance (σ_S): {sigma_S:.4f}")

    return sigma_S, N_S, N_B
```
